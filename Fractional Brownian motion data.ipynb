```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
def generate_fbm(length, hurst):
    fbm = np.zeros(length + 1)
    dt = 1.0 / length
    std_dev = np.sqrt(0.5 * (np.power(2.0, 2.0 * hurst) - 1.0))
    for i in range(1, length + 1):
        fbm[i] = fbm[i - 1] + np.random.normal(scale=std_dev) * np.sqrt(dt)
    return fbm[1:]

def generate_fbm_with_no_noise(length, hurst):
    fbm = generate_fbm(length, hurst)
    fbm += np.random.normal(scale=0, size=length)
    return fbm

def generate_fbm_with_small_noise(length, hurst):
    fbm = generate_fbm(length, hurst)
    fbm += np.random.normal(scale=0.2, size=length)
    return fbm

def generate_fbm_with_middle_noise(length, hurst):
    fbm = generate_fbm(length, hurst)
    fbm += np.random.normal(scale=0.5, size=length)
    return fbm

def generate_fbm_with_big_noise(length, hurst):
    fbm = generate_fbm(length, hurst)
    fbm += np.random.normal(scale= 0.9, size=length)
    return fbm


length = 100  
hurst = 0.7  
```

```python
generate = generate_fbm_with_no_noise(length, hurst)
plt.plot(generate, label='шум 0')
plt.legend()
plt.title('Fractional Brownian motion')
plt.xlabel('Time')
plt.ylabel('Position')
plt.show()
```

```python
generate1 = generate_fbm_with_small_noise(length, hurst)
plt.plot(generate1, label='шум 0.2')
plt.legend()
plt.title('Fractional Brownian motion')
plt.xlabel('Time')
plt.ylabel('Position')
plt.show()
```

```python
generate2 = generate_fbm_with_middle_noise(length, hurst)
plt.plot(generate2, label='шум 0.5')
plt.legend()
plt.title('Fractional Brownian motion')
plt.xlabel('Time')
plt.ylabel('Position')
plt.show()
```

```python
generate3 = generate_fbm_with_big_noise(length, hurst)
plt.plot(generate3, label='шум 0.9')
plt.legend()
plt.title('Fractional Brownian motion')
plt.xlabel('Time')
plt.ylabel('Position')
plt.show()
```

Now we save everything into directory 

```python
import os
```

```python
def generate_fbm_images(length, hurst, noise_scale, num_images, folder_path):
    for i in range(num_images):
        fbm = generate_fbm(length, hurst)
        fbm += np.random.normal(scale=noise_scale, size=length)
        plt.figure(figsize=(6, 4))
        plt.plot(fbm)
        plt.axis('on')  
        plt.savefig(os.path.join(folder_path, f'{i}.png'))
        plt.close()
```

```python
def generate_dataset(length, hurst, noise_levels, num_images_train, num_images_test, num_images_val):
    for noise_scale in noise_levels:
        for dataset_type in ['train', 'test', 'val']:
            folder_path = f'D:/Brownian motion noise dataset hurst_0.7/class_{int(noise_levels.index(noise_scale))}/{dataset_type}'
            os.makedirs(folder_path, exist_ok=True) 
            generate_fbm_images(length, hurst, noise_scale, eval(f'num_images_{dataset_type}'), folder_path)
```

```python

length = 100  
hurst = 0.7  
noise_levels = [0, 0.2, 0.5, 0.9]  
num_images_train = 1000  
num_images_test = 200  
num_images_val = 100  
```

```python
generate_dataset(length, hurst, noise_levels, num_images_train, num_images_test, num_images_val)
```
