```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np
```

```python
desired_width = 100
desired_height = 100

transform = transforms.Compose([
    transforms.Resize((desired_width, desired_height)),  # Замените desired_width и desired_height на желаемые значения
    transforms.ToTensor(),
])

train_set = torchvision.datasets.ImageFolder(root= r'D:\Brownian motion noise dataset', transform=transform)
train_loader = torch.utils.data.DataLoader(train_set, batch_size=4, shuffle=True)
test_set = torchvision.datasets.ImageFolder(root=r'D:\Brownian motion noise dataset', transform=transform)
test_loader = torch.utils.data.DataLoader(test_set, batch_size=4, shuffle=False)
```

```python
class_names = train_set.classes
```

```python
# Функция для отображения изображений и их меток
def image_shower(images, labels, class_names):
    num_images = len(images)
    rows = int(np.ceil(num_images / 4))
    fig, axs = plt.subplots(rows, 4, figsize=(15, 5 * rows))
    for i in range(rows):
        for j in range(4):
            idx = i * 4 + j
            if idx < num_images:
                img = images[idx] / 2 + 0.5
                np_img = img.numpy()
                if rows == 1:
                    axs[j].imshow(np.transpose(np_img, (1, 2, 0)))
                    axs[j].set_title(class_names[labels[idx]])
                    axs[j].axis('off')
                else:
                    axs[i, j].imshow(np.transpose(np_img, (1, 2, 0)))
                    axs[i, j].set_title(class_names[labels[idx]])
                    axs[i, j].axis('off')
    plt.show()

# Визуализация изображений и их меток
images, labels = next(iter(train_loader))
image_shower(images, labels, class_names)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
model = torchvision.models.resnet18(pretrained=True)
```

```python
# Create an instance of the neural network
num_classes = len(class_names)
model.fc = nn.Linear(model.fc.in_features, num_classes)
model = model.to(device)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.1)
num_epochs = 5
```

```python
for epoch in range(num_epochs):
    for i, data in enumerate(train_loader, 0):
        inputs, labels = data[0].to(device), data[1].to(device)

        # Zero the parameter gradients
        optimizer.zero_grad()

        # Forward pass
        outputs = model(inputs)
        loss = criterion(outputs, labels)

        # Backward pass and optimization
        loss.backward()
        optimizer.step()

        # Print statistics
        if i % 100 == 0:
            print(f"Epoch {epoch + 1}, Iteration {i + 1}, Loss: {loss.item()}")

print("Training finished!")
```

```python
model.eval()  # Установка модели в режим оценки
correct = 0
total = 0
```

```python
import random


with torch.no_grad():
    examples_per_class = 2  # Количество примеров из каждого класса для отображения
    total = 0
    correct = 0

    # Получение индексов изображений для каждого класса
    class_indices = [[] for _ in range(len(class_names))]
    for i, (images, labels) in enumerate(test_loader):
        for j, label in enumerate(labels):
            class_indices[label].append(i * test_loader.batch_size + j)

    # Выбор случайных примеров из каждого класса и их отображение
    for class_index in class_indices:
        random.shuffle(class_index)
        class_index = class_index[:examples_per_class]  # Берем только examples_per_class случайных изображений из каждого класса
        for idx in class_index:
            images, labels = test_set[idx]
            images = images.unsqueeze(0).to(device)
            labels = torch.tensor(labels).unsqueeze(0).to(device)  # Преобразуем labels в тензор
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)

            image = images.squeeze() / 2 + 0.5  # Unnormalize
            np_image = image.permute(1, 2, 0).cpu().numpy()  # Convert to numpy array
            plt.imshow(np_image)
            plt.title(f"Predicted: {class_names[predicted.item()]}, True: {class_names[labels.item()]}")
            plt.axis('off')
            plt.show()

            total += 1
            correct += (predicted == labels).item()

    accuracy = 100 * correct / total
    print(f"Accuracy on the entire test set: {accuracy:.2f}%")
```
